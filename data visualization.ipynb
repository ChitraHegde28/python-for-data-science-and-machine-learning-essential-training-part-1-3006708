```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame,Series
from matplotlib import rcParams
import pandas as pd
import numpy as np
```

```python
df=DataFrame({ 'name':['steve','mike','luke','will','dustin','max','el'],
      'age':[25,15,16,17,18,19,18],
      'gender':['male','male','male','male','male','female','female'],
      'rank':[1,2,1,1,2,2,1],
      })
df
```

# **subplot**

```python
fig=plt.figure()
fig ,axes=plt.subplots(nrows=2,ncols=2)  #subplot (1,2) 1-row,2-columns
(ax1,ax2),(ax3,ax4)=axes
ax1.plot(x, y, c='red')
ax2.plot(x, y, c='red')
ax3.plot(x, y, c='blue')
ax4.plot(x, y, c='blue')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0, 0].plot(x, y, c='red')
ax[0, 1].plot(x, y, c='red')
ax[1, 0].plot(x, y, c='blue')
ax[1, 1].plot(x, y, c='blue')
```

# object oriented method

```python
address='/workspaces/jupyter-example/mtcars.csv'
cars=pd.read_csv(address)
cars.columns=['cars_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
mpg=cars['mpg']
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.cars_names,rotation=60,fontsize='medium')
ax.set_title('mpg of cars')
ax.set_xlabel('carnames')
ax.set_ylabel('mpg')
```

**add_axes method is used to add an axis to the figure at the position specified by the list [0.1, 0.1, 0.8, 0.8]. This means the new axis starts at 10% from the left and bottom of the figure and spans 80% of the figure's width and height.**

```python

fig=plt.figure()
ax=fig.add_axes([0.1,0.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.cars_names,rotation=50,fontsize='medium')
ax.set_title('mpg of cars')
ax.set_xlabel('carnames')
ax.set_ylabel('mpg')
ax.legend(loc='best')
```
