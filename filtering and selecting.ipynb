# Filtering and Selecting

**missing values 
fillna(),replace(),interpolate()**

```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
data={'name':['richard','millie','mike','max','luke','dustin','steve'],
      'age':[20,22,23,23,25,26,22],
      'gender':['male','female','male','female','male','male','male'],
       'rank':[1,2,1,3,2,2,3]
      }
ranking_df=DataFrame(data)
ranking_df
```

```python
ranking_df.iloc[2:4,1]=np.nan
ranking_df.iloc[3:6,3]=np.nan
ranking_df.iloc[1:3,2]=np.nan
ranking_df
```

```python
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

```python
bool_series=pd.isnull(ranking_df['gender'])
bool_series
ranking_df[bool_series]
```

**fill the NaN with the previous by passing method='pad' or method='ffill'
fill the NaN with the next by passing method='bfill'**

```python
ranking_df.fillna(method='ffill')
```

```python
ranking_df.fillna(method='bfill')
```

```python
ranking_df.interpolate(method='linear')
```

```python
ranking_df
```

```python
ranking_df.dropna(how='all')
```

axis=1 means the column which has  NaN is droped
axis=0 specifys rows

```python
ranking_df.dropna(axis=1)
```
