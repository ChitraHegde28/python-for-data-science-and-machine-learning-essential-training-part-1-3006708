```python
Q1 = df['Sepal Width'].quantile(0.25)
Q3 = df['Sepal Width'].quantile(0.75)
IQR = Q3 - Q1
Q1
Q3
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
lower_bound
upper_bound
```

```python
outliers = df[(df['Sepal Width'] < lower_bound) | (df['Sepal Width'] > upper_bound)]
print("Outliers:\n", outliers)
```
