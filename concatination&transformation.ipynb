```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
df_obj=DataFrame(np.arange(36).reshape(6,6))
df_obj
```

```python
df_obj1=DataFrame(np.arange(15).reshape(3,5))
df_obj1
```

```python
pd.concat([df_obj,df_obj1],axis=1)
```

```python
#droping data
df_obj.drop([0,2])  #droping row
```

```python
df_obj.drop([0,2],axis=1) #droping column
```
